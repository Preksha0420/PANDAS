```python
# code to identify outliers.
def find_and_display_outliers(dataframe, column_name, additional_column):
    """
    This function finds and displays the outliers in a specific column of a dataframe using the IQR method.
    params:
        - dataframe: pandas DataFrame
        - column_name: Name of the column for which outliers are to be calculated
        - additional_column: Additional column to be displayed along with 'Gnrc_Name' and 'Tot_Spndng_2017'
    returns:
        - A DataFrame containing 'Gnrc_Name', 'Tot_Spndng_2017', and the specified additional column for rows with outliers
    """
    # extracting the specified column
    pd_object = dataframe[column_name]

    # checking if the column contains numerical data
    if not pd.api.types.is_numeric_dtype(pd_object):
        raise ValueError(f"The column '{column_name}' does not contain numerical data.")

    # calculating Q1 (also known as 25th percentile)
    q1 = pd_object.quantile(0.25)

    # calculating Q3 (also known as 75th percentile)
    q3 = pd_object.quantile(0.75)

    # calculating the IQR
    iqr = q3 - q1

    # calculating the upper and lower bounds
    upper = q3 + 1.5 * iqr
    lower = q1 - 1.5 * iqr

    # finding the outliers
    outliers_mask = (pd_object < lower) | (pd_object > upper)

    # extracting the rows with outliers
    outliers_df = dataframe[outliers_mask][['Gnrc_Name', column_name, additional_column]]

    return outliers_df
```

```python
# Finding outliers for 2021
outliers_result = find_and_display_outliers(not_overall, 'Tot_Spndng_2021', 'Mftr_Name')
# percentage of outliers in the year 2021
perc = (outliers_result.Tot_Spndng_2021.sum()/not_overall.Tot_Spndng_2021.sum()*100).round(2)
print(f' The percentage of outliers in the entire dataset for diabetes medication is {perc} %.')
```

```python
# grouping the data and adding up the spending by Manufacturer name
outliers_result = outliers_result.groupby('Mftr_Name')['Tot_Spndng_2021'].sum()
# sort the data
outliers_result = outliers_result.sort_values(ascending=False)
```

```python
# selecting the top 5 results
top_5_outliers = outliers_result.head(5)

# creating a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=top_5_outliers.values, y=top_5_outliers.index, palette='viridis')

# labels and title
plt.xlabel('Total Spending in 2021')
plt.ylabel('Manufacturer Name')
plt.title('Top 5 Manufacturers - Total Spending in 2021')

# showing the plot
plt.show()
```

```python
outlier_result = find_and_display_outliers(not_overall, 'Tot_Spndng_2021', 'Mftr_Name')

outlier_result = outlier_result.groupby(['Gnrc_Name', 'Mftr_Name'])['Tot_Spndng_2021'].sum()
```

```python
# reset index to make 'Gnrc_Name' and 'Mftr_Name' columns
outlier_result = outlier_result.reset_index()

# pivoting the data for the heatmap
heatmap_data = outlier_result.pivot('Gnrc_Name', 'Mftr_Name', 'Tot_Spndng_2021')

# making a heatmap with grid lines
plt.figure(figsize=(15, 10))
sns.heatmap(heatmap_data, cmap='plasma', fmt=".2e", linewidths=1, linecolor='white', cbar_kws={'label': 'Total Spending'})
plt.title('Total Outlier Spending in 2021 - Diabetic Medications')
plt.xlabel('Manufacturer Name')
plt.ylabel('Generic Name')
plt.show()
```

```python
# select top N manufacturers and generic names
top_n = 10
top_manufacturers = outlier_result.groupby('Mftr_Name')['Tot_Spndng_2021'].sum().nlargest(top_n).index
top_generic_names = outlier_result.groupby('Gnrc_Name')['Tot_Spndng_2021'].sum().nlargest(top_n).index

# filter data for heatmap
filtered_heatmap_data = outlier_result[outlier_result['Mftr_Name'].isin(top_manufacturers) & outlier_result['Gnrc_Name'].isin(top_generic_names)]
pivot_table = filtered_heatmap_data.pivot(index='Gnrc_Name', columns='Mftr_Name', values='Tot_Spndng_2021')

# create the enhanced heatmap
plt.figure(figsize=(10, 10))
sns.heatmap(pivot_table, cmap='coolwarm', annot=True, fmt=".2e", linewidths=.5, linecolor='black', cbar_kws={'label': 'Total Spending'})
plt.title('Top Manufacturers vs Generic Names - Total Spending in 2021')
plt.show()
```

```python
# simplified stacked bar chart for top N generic drugs
top_n_generic = 5
top_generic_drugs = outlier_result.groupby('Gnrc_Name')['Tot_Spndng_2021'].sum().nlargest(top_n_generic).index
filtered_data = outlier_result[outlier_result['Gnrc_Name'].isin(top_generic_drugs)]

# pivot for stacking
pivot_data = filtered_data.pivot_table(index='Gnrc_Name', columns='Mftr_Name', values='Tot_Spndng_2021', aggfunc='sum', fill_value=0)

# plotting the stacked bar chart
pivot_data.plot(kind='bar', stacked=True, figsize=(10, 5))
plt.title('Spending Distribution Across Manufacturers for Top Generic Drugs in 2021')
plt.ylabel('Total Spending')
plt.xlabel('Generic Drug Name')
plt.legend(title='Manufacturer', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

#### Insights

As we see above, **97.42%** of the total spending is on the outliers and hence, this needs to be controlled and checked.

Further looking into reasons behind this, as we can see the **top 5 companies** that are responsible for outlier spending patterns also have a monopoly in the drugs they create and hence charge exceptionally high amounts. This is an opportunity for new pharma companies or pharmaceutical companies finding an opportunity to dive into.

Moreover, insulin is the prime drug for diabetes and also responsible for the most spending within outliers. This si one area to cap the prices within drug category.

Also, Medicare needs to negotiate its terms with these manufacturers in ways using policy reform, negotiating the prices, and leveraging the last customer base they have.

Also, the companies that manufacture outlier drugs at lower costs and fall in the normal range, can be looked into for negotiation and providing them with better incentives.

2. **Outlier Spending on Specific Drugs:**

3. **Trend of Increasing Spending:**

4. **Impact of Monopolistic Practices:**
    - Certain manufacturers with monopolies over specific drugs tend to increase prices, which contributes to the outlier spending





#### Recommendations for Stakeholders
1. **Negotiation of Drug Prices:**
    - We recommend Medicare negotiate directly with drug manufacturers to secure better prices, specifically for high-cost diabetic medications that are outliers in terms of spending cost.
2. **Encourage Generic and Biosimilar Drug Use:**
    - We recommend Medicare to promote the use of generic and biosimilar versions of expensive diabetic medications to provide a more cost-effective alternative without compromising treatment quality.
3. **Implement Price Caps:**
    - We recommend setting price caps for essential diabetic medications to prevent price gouging and ensure affordability for Medicare beneficiaries.
4. **Supporting Pharmaceutical Competition:**
    - We recommend encouraging competition in the pharmaceutical market, in particular for the drugs that are currently monopolized, which will lower prices and increase access to those diabetic medications.



#### Conclusion
